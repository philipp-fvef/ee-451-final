# Image cropping

This script takes the four reference images from data/reference_images
and crops them such that there is exactly one card per image
and saves the cropped images in data/reference_images 
with the card visible as the file name, e.g. "y_2.jpg" or "wild.jpg".

```python
reference_image_1_path = "data/reference_images/L1000765.jpg"
reference_image_2_path = "data/reference_images/L1000766.jpg"
reference_image_3_path = "data/reference_images/L1000767.jpg"
reference_image_4_path = "data/reference_images/L1000768.jpg"
```

## Reference Image 1

```python
# display image and define crop lines
import cv2
import matplotlib.pyplot as plt
from numpy import mean

reference_image_1 = cv2.imread(reference_image_1_path)
reference_image_1 = cv2.cvtColor(reference_image_1, cv2.COLOR_BGR2RGB) 

# create a dictionary of card names and their outlines (y min y max x min x max) for cropping
card_crop_lines_ref_1 = {
    "y_1": {"x": [1250, 1650], "y": [630, 1230]},
    "y_2": {"x": [1270, 1670], "y": [1280, 1860]},
    "y_3": {"x": [1280, 1680], "y": [1880, 2470]},

    "r_1": {"x": [1680, 2100], "y": [640, 1230]},
    "r_2": {"x": [1710, 2100], "y": [1290, 1860]},
    "r_3": {"x": [1730, 2120], "y": [1890, 2470]},

    "b_1": {"x": [2120, 2510], "y": [630, 1220]},
    "b_2": {"x": [2120, 2530], "y": [1270, 1840]},
    "b_3": {"x": [2160, 2580], "y": [1880, 2460]},

    "g_1": {"x": [2540, 2940], "y": [630, 1210]},
    "g_2": {"x": [2550, 2940], "y": [1250, 1830]},
    "g_3": {"x": [2580, 2970], "y": [1880, 2450]},
}

# show the outlines and labels on the image
plt.imshow(reference_image_1)
for card_name, crop_lines in card_crop_lines_ref_1.items():
    vertical = crop_lines["y"]
    horizontal = crop_lines["x"]
    plt.plot(horizontal, [vertical[0], vertical[0]], color="red")
    plt.plot(horizontal, [vertical[1], vertical[1]], color="red")
    plt.plot([horizontal[0], horizontal[0]], vertical, color="red")
    plt.plot([horizontal[1], horizontal[1]], vertical, color="red")
    plt.text(mean(horizontal) - 100, mean(vertical), card_name, color="black", fontsize=12)
plt.show()
```

```python
# export the cropped images
for card_name, crop_lines in card_crop_lines_ref_1.items():
    vertical = crop_lines["y"]
    horizontal = crop_lines["x"]
    cropped_image = reference_image_1[vertical[0]:vertical[1], horizontal[0]:horizontal[1]]
    cv2.imwrite(f"data/reference_images/{card_name}.jpg", cv2.cvtColor(cropped_image, cv2.COLOR_RGB2BGR))
```
